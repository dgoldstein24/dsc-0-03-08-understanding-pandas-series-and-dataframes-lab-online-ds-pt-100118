## Create another column 'Num_Lines' that is a count of how many lines pass through a station. Then sort your dataframe by this column in descending order.

```python
stations_and_lines = df.groupby('station')['linename'].value_counts().to_dict()
pairs = list(stations_and_lines.keys())
pairs
stations = []
for pair in pairs:
    stations.append(pair[0])
stations
unique_stations = list(set(stations))
unique_stations

new_column = {}


for unique_station in unique_stations:
    count = 0
    for station in stations:
        if unique_station == station:
            count += 1
    new_column.update({unique_station : count})

df['Num_Lines'] = df.station.map(new_column)
df.sort_values(by = 'Num_Lines', ascending = False)


    
    
```

## Compare subway traffic by day of the week. Display this as a graph.

```python
df.head()
dates = pd.to_datetime(df.Date)
DoW = dates.dt.day_name()
DoW
df['DoW'] = DoW

DoW_freq = df.groupby('DoW')['Entries'].mean()
x = list(DoW_freq.keys())
y = [DoW_freq[x[0]], DoW_freq[x[1]], DoW_freq[x[2]], DoW_freq[x[3]], DoW_freq[x[4]], DoW_freq[x[5]], DoW_freq[x[6]]]


plt.bar(x, y, label = 'Traffic')
plt.title('Traffic by DoW')
plt.legend()
plt.show()
```

## Is there more subway traffic on a weekend or a weekday?    Be specific in comparing magnitudes.

```python
#Your code here

weekend_values = y[2:4]
weekend_total = 0
for value in weekend_values:
    weekend_total += value
weekend_avg = weekend_total / len(weekend_values)


weekday_values = [y[0],y[1],y[4],y[5],y[6]]
weekday_total = 0
for value in weekday_values:
    weekday_total += value
weekday_avg = weekday_total / len(weekday_values)

if weekend_avg > weekday_avg:
    print("weekend has more")
elif weekend_avg == weekday_avg:
    print("same")
elif weekend_avg < weekday_avg:
    print("weekday has more")

# There is more traffic on the weekend
```
